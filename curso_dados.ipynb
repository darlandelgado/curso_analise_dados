#**Obtenção dos dados**

##Existem várias fontes para obter dados e informações:
- Dados publicados pelo governo, indústria ou indivíduos;
- Dados oriundos de experiências (experimentos);
- Dados oriundos de pesquisa;
- Dados oriundos de observações de comportamentos, atitudes etc.

#**Dados primários**: São informações e dados coletados pela primeira vez por um pesquisador para atender a um propósito específico de pesquisa (problema de pesquisa e seus objetivos). Características:
- **_Originalidade_**: coletados diretamente da fonte de ocorrência, como em pesquisas, experimentos (laboratoriais ou sociais), entrevistas e outras formas de contato direto com os fatos observados pelo pesquisador.
- **_Especificidade_**: atende exatemente às necessidades metodológicas da pesquisa concebida pelo pesquisador.
- **_Precisão_**: controle sobre obtenção/acesso dos dados, confiabilidade e qualidade.

#**Dados secundários**: São informações e dados já existentes, coletados e organizados por outras pessoas/entidades (institutos de pesquisa, governo, indústria, entre outras)para atender a um propósito desejado. Geralmente, são criadas bases de dados em séries históricas para acesso por terceiros. Características:
- **_Preexistentes_**: coletados de fontes externas, como livros, artigos científicos, artigos de revistas, relatórios, anuários estatísticos, base de dados e websites, entre outras formas.
- **_Organizados_**: foram estruturados, organizados e tabulados por outras pessoas, geralmente pesquisadores e/ou profissionais interessados nas informações e dados da temática objeto da base elaborada.
- **_Acessíveis_**: estão disponíveis para acesso público, de forma gratuita ou paga.
- **_Confiáveis_**: geralmente são dados categorizados, filtrados estatisticamente e processados com metodologia rigorosa. Fontes podem ser garantia de fidedignidade dos dados.

Na abordagem abaixo, usamos a função str.extract() com a expressão regular r'(\d+\.\d+)', que corresponde a qualquer sequência de dígitos seguida por um ponto e mais dígitos, representando um número decimal. Isso garante que apenas os valores que seguem esse padrão sejam extraídos da coluna "Índice". Em seguida, convertemos esses valores extraídos para o tipo float.

```python
# Filtra os valores que correspondem ao padrão de índice e converte para float
tabela2['Índice'] = tabela2['Índice'].str.extract(r'(\d+\.\d+)', expand=False).astype(float)
tabela2
```

```python
# O método "describe" exibe as principais medidas estatísticas
medidas = tabela2['Índice'].describe()
medidas
```

```python
# Calcula a amplitude da coluna 'Índice'
amplitude = tabela2['Índice'].max() - tabela2['Índice'].min()
amplitude
```

```python
# Obtem de "medidas" os valores desejados individualmente
media = medidas['mean'].round(2)
dp = medidas['std'].round(2)
cv = dp / media

# Imprime na tela os valores da media, dp e cv
print(f"media = {media}, dp = {dp} e cv = {cv}")
```

Observe, no DataFRame de SP, que a coluna "Safra" está ordenada de modo decrescente, ou seja, da safra mais atual para a mais antiga. Isso não é adequado para se fazer um gráfico de série temporal.

Vamos alterar (inverter) a ordem.

```python
# Inverte a ordem das linhas no DataFrame tabelaSP
tabelaSP = tabelaSP.iloc[::-1]
```

```python
tabelaSP
```

Vamos elaborar um gráfico de linha que evidencie a dinâmica da produção de Açúcar e de Etanol no período compreendido no DataFrame (safra 2000/2001 até safra 2022/2023.

```python
import matplotlib.pyplot as plt

# Criar uma figura e um eixo
fig, ax = plt.subplots(figsize=(10, 6))

# Plotar os dados da produção de açúcar
ax.plot(tabelaSP['SAFRA'], tabelaSP['AÇÚCAR'], label='Açúcar')

# Plotar os dados da produção de etanol total
ax.plot(tabelaSP['SAFRA'], tabelaSP['ETANOL TOTAL'], label='Etanol Total')

# Adicionar título e rótulos dos eixos
ax.set_title('Produção de Açúcar e Etanol em SP - 2000-2023')
ax.set_xlabel('Safras')
ax.set_ylabel('Produção')

# Rotacionar os rótulos do eixo x em 45 graus
plt.xticks(rotation=45)

# Adicionar legenda
ax.legend()

# Exibir o gráfico
plt.show()
```

Para calcular as variações percentuais ($\Delta$%) de uma dada safra para a próxima no DataFrame tabelaSP, pode-se usar o método pct_change() do Pandas.

Observe abaixo o código referente à produção de Açúcar:

```python
# Calcula as variações percentuais de uma safra para a próxima de Açúcar
# Cria uma coluna para exibir os percentuais de variação
tabelaSP['Var Perc AÇÚCAR'] = tabelaSP['AÇÚCAR'].pct_change() * 100
```

```python
tabelaSP
```

Vamos calcular a média (media_A) de produção de **Açúcar** por safra, no período do DataFrame, o desvio-padrão (dp_A) e o coeficiente de variação de Pearson (CV_a).

```python
# Cálculo da média anual da produção de Açúcar:
media_A = tabelaSP['AÇÚCAR'].mean().round(2)

# Cálculo do desvio-padrão amostral da produção de Açúcar:
dp_A = tabelaSP['AÇÚCAR'].std().round(2)

# Cálculo do coeficiente de variação de Pearson da produção de Açúcar:
cv_A = dp_A / media_A

# Exibe na tela os valores
print(f"Média de produção de Açúcar = {media_A:,.2f}, dp = {dp_A:,.2f}, CV = {cv_A:,.2f}")
```

Vamos calcular a média (media_A) de produção de **Etanol Total** por safra, no período do DataFrame, o desvio-padrão (dp_A) e o coeficiente de variação de Pearson (CV_a).

```python
# Cálculo da média anual da produção de Etanol:
media_E = tabelaSP['ETANOL TOTAL'].mean().round(2)

# Cálculo do desvio-padrão amostral da produção de Etanol:
dp_E = tabelaSP['ETANOL TOTAL'].std().round(2)

# Cálculo do coeficiente de variação de Pearson da produção de Etanol:
cv_E = dp_E / media_E

# Exibe na tela os valores
print(f"Média de produção de Etanol = {media_E:,.2f}, dp = {dp_E:,.2f}, CV = {cv_E:,.2f}")
```
